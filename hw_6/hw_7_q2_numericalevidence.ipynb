```python
#libraries
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
from scipy.optimize import minimize
```

```python
A = np.array([[2,1,0],[1,2,0],[0,0,2]])
B = np.array([[2,-1,0],[-1,2,0],[0,0,2]])
mean1 = np.array([2,0,0])
mean2 = -mean1
```

```python
B
```

```python
xval = np.linspace(-6,6,100)
x1,x2,x3 = np.meshgrid(xval,xval,xval)
```

```python
val = np.zeros((100,100,100))
contour = np.zeros(3)
for i in range(100):
    for j in range(100):
        for k in range(100):
            x = np.array([x1[i,j,k],x2[i,j,k],x3[i,j,k]])
            val[i,j,k] = -0.5*(x - mean1).T @ A @ (x - mean1) + 0.5*(x - mean2).T @ B @ (x - mean2)
            if (abs(val[i,j,k]) < 0.3):
                contour = np.column_stack([contour,x])
```

```python
d1 = np.random.multivariate_normal(mean1, np.linalg.inv(A), 500)
d2 = np.random.multivariate_normal(mean2, np.linalg.inv(B), 500)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(d1[:,0],d1[:,1],d1[:,2],s=10, color="orange")
ax.scatter(d2[:,0],d2[:,1],d2[:,2],s=10, color="blue")
ax.scatter(contour[0,:],contour[1,:],contour[2,:],color="black")
ax.view_init(0, 0)
```
